### Solution ###

```python
tq = estt.TimeQuery("data.ts", 1469073320.448689, 1469073905.999965)
print tq.get_query()
locs_df = esda.get_data_df("background/filtered_location", curr_id, time_query=tq)
```

```python
locs_df[["fmt_time", "ts", "latitude", "longitude", "metadata_write_ts"]]
```

```python
locs_df.ts.diff()
```

```python
diff_ts = locs_df.ts.diff()
diff_ts[diff_ts > 60].index
```

```python
for i in diff_ts[diff_ts > 60].index:
    print i
```

```python
ecc.calDistance(locs_df.iloc[1]["loc"]["coordinates"], locs_df.iloc[2]["loc"]["coordinates"])
```

```python
ts_fill = np.arange(locs_df.ts[1] + 60, locs_df.ts[2], 60)
```

```python
ts_fill[0:10]
```

```python
dist_fill = np.random.uniform(low=0, high=100, size=len(ts_fill))
```

```python
angle_fill = np.random.uniform(low=0, high=2*np.pi, size=len(ts_fill))
```

Formula from 
http://gis.stackexchange.com/questions/5821/calculating-latitude-longitude-x-miles-from-point

```python
lat_fill = locs_df.latitude[2] + np.multiply(dist_fill, np.sin(angle_fill)/111111)
```

```python
cl = np.cos(locs_df.latitude[2])
lng_fill = locs_df.longitude[2] + np.multiply(dist_fill, np.cos(angle_fill)/cl/111111)
```

```python
len(dist_fill), len(angle_fill), len(lat_fill), len(lng_fill)
```

```python
dist_fill[0:10], angle_fill[0:10], lat_fill[0:10], lng_fill[0:10]
```

```python
loc_fill = [gj.Point(l) for l in zip(lng_fill, lat_fill)]
```

```python
loc_fill[0:10]
```

```python
fill_df = pd.DataFrame({"ts": ts_fill, "longitude": lng_fill, "latitude": lat_fill, "loc": loc_fill})
```

```python
len(fill_df)
```

```python
fill_df.head()
```

```python
filled_df = pd.concat([locs_df, fill_df]).reset_index().sort("ts")
```

```python
filled_df[["fmt_time", "ts", "latitude", "longitude", "metadata_write_ts"]]
```

```python
ipy.inline_maps([lo.get_maps_for_geojson_unsectioned([gfc.get_feature_list_from_df(filled_df)])])
```

```python
with_speed_df = eaicl.add_dist_heading_speed(filled_df)
```
